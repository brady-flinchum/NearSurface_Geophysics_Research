# PyVista Example

This notebook demonstrates an interactive 3D figure embedded in the site.
In CI, PyVista renders to a static screenshot. Locally, swap the backend to `trame` for full interactivity.

````python
import pyvista as pv
import numpy as np

# Use static backend for reproducible builds.
# Locally: pv.set_jupyter_backend('trame')
pv.set_jupyter_backend('static')
pv.OFF_SCREEN = True

# --- Example: warped surface ---
mesh = pv.Sphere(radius=1.0, theta_resolution=60, phi_resolution=60)

# Add a scalar field (e.g. distance from a point)
center = np.array([0.5, 0.5, 0.5])
distances = np.linalg.norm(mesh.points - center, axis=1)
mesh['distance'] = distances

pl = pv.Plotter(notebook=True)
pl.add_mesh(mesh, scalars='distance', cmap='viridis',
            show_scalar_bar=True, scalar_bar_args={'title': 'Distance'})
pl.add_axes()
pl.show()
````

## Explanation

Replace the mesh above with your own simulation output. You can load VTK files with:

```python
mesh = pv.read('path/to/your/file.vtk')
```

For interactive figures in the browser, use `pv.set_jupyter_backend('trame')` locally,
but keep `'static'` in the notebook you commit so CI builds work.
